```python
email_senha = {
    'Login123': {
        'senha': '12345',
        'funcao': 'Cliente',
        'pedido': [],
        'visto': 'Não'
    },
    '123': {
        'senha': '123',
        'funcao': 'Empregado',
        'pedido': []
    }
}
#cardapio = {Numero_pedido: [Nome_prato, Preço]}
cardapio = {
    '1': ["PF Bife Acebolado", 24.90],
    '2': ["PF Frango Grelhado", 22.90],
    '3': ["PF Ovo Frito", 17.90],
    '4': ["Feijoada", 27.90],
    '5': ["Estrogonofe de Frango", 25.90],
    '6': ["Macarronada com Carne Moída", 23.90],
    '7': ["Suco natural", 6.00],
    '8': ["Refrigerante lata", 5.00],
    '9': ["Água mineral", 3.00],
    '10': ["Pudim de leite", 6.00],
    '11': ["Gelatina de sabores", 4.00],
    '12': ["Salada de frutas", 6.00]}

senha_admin = 'admin_123'

#Salvamento de dados em um arquivo local
import json

#Função para salvar dados em cadastro.json
def salvar_dados():
    """cadastro.json"""
    with open('cadastro.json', 'w') as f:
        json.dump(email_senha, f, indent=4)

#Abre o arquivo cadastro.json para salvar as informações de cadastro 
try:
    with open('cadastro.json', 'r') as f:
        email_senha.update(json.load(f))
#Cria o arquivo cadastro.json se ele ainda não existe ou foi deletado 
except FileNotFoundError:
    print('Arquivo não encontrado. Criando novo cadastro.json')
    salvar_dados()

#Função para salvar o cardapio
def salvar_cardapio():
    '''cardapio.json'''
    with open('cardapio.json', 'w') as f:
        json.dump(cardapio, f, indent=4)
#Cria um arquivo cardapio.json, para salvar as informações alteradas do cardápio. 
try:
    with open('cardapio.json', 'r') as f:
        cardapio.update(json.load(f))
except FileNotFoundError:
    salvar_cardapio()

def login():
    print('Login...')
    while True:
        email = input('Usuário: (ou sair) ').capitalize().strip()

        if email == 'Sair':
            return None
        
        senha = input('Senha: ')

        #Checagem se o usuario está dentro do dicionario(email_senha) e se a senha está correta 
        if email in email_senha and email_senha[email]['senha'] == senha:
            print(f'Acesso concedido...\n{email} logado com sucesso.')
            return email, email_senha[email]['funcao']
        else:
            print('Usuário ou senha incorretos...\nTente novamente')
            continue

def cadastro():
  print('Cadastro...')
  while True:
    #Opção de login ou cadastro
    conta = input('Já possui um login? (s/n) ').lower().strip()
    if conta == 's':
        return login()

    elif conta == 'n':
        email = input('Novo usuario: (ou sair)').capitalize().strip()

        if email == 'Sair':
            return None
        
        #Verificação se o usuario já existe
        if email in email_senha:
            print(f'O usuario {email} já foi cadastrado\nFaça o login.')
            return None
            
        senha = input('Senha: ')
        
        #Usuario é Cliente ou Empregado
        print('A opção a seguir não poderá ser alterada, responda com atenção...')
        while True:
            funcao = input('Cliente ou Empregado: ').capitalize().strip()
            if funcao == 'Cliente':
                email_senha[email] = {
                        'senha': senha,
                        'funcao': funcao,
                        'pedido': [],
                        'visto': 'Não'
                    }
                print(f'Cadastro de {funcao} realizado com sucesso!')
                salvar_dados()
                return email, funcao
            #Verificação de Funcionário
            elif funcao == 'Empregado':
                input_senha = input('Senha de Empregado (ou sair): ')
                if input_senha == 'sair':
                    print('Se você não é funcionário, digite Cliente.')
                    continue
                if input_senha == senha_admin:
                    print(f'Cadastro de {funcao} realizado com sucesso!')
                    email_senha[email] = {
                        'senha': senha,
                        'funcao': funcao,
                        'pedido': []
                    }
                    salvar_dados()
                    return email, funcao
                else:
                    print('Senha incorreta...\nSe você não é funcionário, digite Cliente.')
            else:
                print('Digite uma opção válida -> (Cliente ou Empregado)')
    else:
        print('Opção inválida...\nDigite "s" ou "n".')


def pedido():
    for index, (nome_prato, preco) in cardapio.items():
        print(f'{index} {nome_prato} - R${preco:.2f}')
    preco_total = 0
    pedido = []
    while True:
        try:
            print("Digite o número dos itens do cardápio um por um.")
            num_pedido = input('Número do pedido (0 para encerrar): ')
            
            if num_pedido == '0':
                print('Encerrando pedido...')
                break
            #Adiciona o nome e preço a lista do pedido
            if num_pedido in cardapio:
                nome, preco = cardapio[num_pedido]
                preco_total += preco
                pedido.append([nome, preco])
                print(f'O pedido {nome} foi anotado')
                continue

            else:
                print(f'O pedido solicitado não está na lista')
                continue

        except ValueError:
            print(f'Digite um número válido...\nEntre 1 e {len(cardapio)}')
    if pedido == []:
        print('Nenhum item foi pedido...')
        return None
    else:
        print('Resumo do Pedido')
        #Imprime a lista de pedido um por um; com o preço total ao final
        for nome, preco in pedido:
            print(f'{nome} - R${preco}')
        print(f'Total a pagar: R${preco_total}')
        return pedido

def gerenciar_pedidos():
    '''Checa os pedidos já feitos e atualiza o status para "visto" ou "entregue"'''
    print('--- Lista de Pedidos dos Clientes ---')
    continuar = input('Deseja prosseguir (s/n)? ').lower().strip()
    if continuar == 's':    
        pedidos_encontrados = False

        for usuario, dados in email_senha.items():
            if dados['funcao'] == 'Cliente' and dados['pedido']:
                status_atual = dados['visto']
                pedidos_encontrados = True
                print(f'\nCliente: {usuario} - Status do pedido: {status_atual}') 
                while True:
                    acao = input('Deseja marcar como (visto/entregue)? ').lower().strip()

                    if acao == 'visto':
                        email_senha[usuario]['visto'] = 'Sim'  #Marca o pedido como visto
                        print('Pedido marcado como visto.')
                        break
                    elif acao == 'entregue':
                        email_senha[usuario]['pedido'] = []   #Limpa a lista de pedidos
                        email_senha[usuario]['visto'] = 'Não'  #Zera o status
                        print('Pedido marcado como entregue e removido.')
                        break
                    else:
                        print('Opção inválida. Digite "visto", "entregue" ou "pular".')

        if not pedidos_encontrados:
            print('Nenhum pedido encontrado.')

        salvar_dados() 
    else:
        print('Fechando função...')
        return None

def adicionar_pratos():
    print('--- Adicionar ao Cardápio ---')
    continuar = input('Deseja prosseguir (s/n)? ').lower().strip()
    if continuar == 's':
        while True:
            n_prato = len(cardapio) + 1
            novo_prato = input('Nome do novo Prato (ou "sair"): ').strip().capitalize()
            if novo_prato == 'Sair':
                break
            try:
                preco_novo_prato = float(input('Preço (R$): '))
                print(f'{n_prato}: [{novo_prato}, {preco_novo_prato}]')
                adicionar = input(f'As informações estão corretas (s/n)? ')
                if adicionar == 's':
                    cardapio.update({str(n_prato): [novo_prato, preco_novo_prato]})
                    salvar_cardapio()
                    break
                else:
                    print('Insira as informações novamente.')
            except ValueError:
                print('Digite um preço válido.')
    else:
        print('Fechando função...')
        return None

def reindexar_cardapio():
    '''Reeorganiza as chaves do cardapio quando um item é removido.'''
    global cardapio
    itens = list(cardapio.values())  # lista só dos valores, em ordem de inserção
    cardapio = {}  # limpa o dicionário atual

    for i, item in enumerate(itens, start=1):
        cardapio[str(i)] = item

def remover_pratos():
    print('--- Remover do Cardápio ---')
    continuar = input('Deseja prosseguir (s/n)? ').lower().strip()
    if continuar == 's':
        while True:
            try:
                print('Digite "sair" para sair.')
                n_prato = input(f'Número do prato a ser removido [1 a {len(cardapio)}]: ')
                if n_prato == 'sair':
                    break
                if n_prato not in cardapio:
                    print('Número inválido. Tente novamente.')
                    continue
                remover = input(f'Remover {cardapio[n_prato][0]} (s/n)? ')
                if remover == 's':
                    print(f'Removendo item {cardapio[n_prato][0]} do cardápio...')
                    cardapio.pop(n_prato)
                    reindexar_cardapio()
                    salvar_cardapio()
                    break
                else:
                    print('O item não será removido.')
                    print('Fechando função...')
                    return None
            except ValueError:
                print('Digite corretamente.')
    else:
        print('Fechando função...')
        return None

def alterar_precos():
    print('--- Alterar preços do Cardápio ---')
    continuar = input('Deseja prosseguir (s/n)? ')
    if continuar == 's':
        while True:
            try:
                print('Digite "sair" para sair.')
                n_prato = input(f'Número do prato para alteração de preço [1 a {len(cardapio)}]: ')
                if n_prato == 'sair':
                    break
                if n_prato not in cardapio:
                    print('Número inválido. Tente novamente.')
                    continue
                preco_item = cardapio[n_prato][1]
                alterar = input(f'O preço a ser alterado é R${preco_item} do item {cardapio[n_prato][0]} (s/n)? ').lower().strip()
                if alterar == 's':
                    desconto = int(input('Qual opção [1 ou 2]? '))
                    if desconto == 1:
                        porcentagem = float(input('Quanto é o desconto (%): '))
                        percento = porcentagem/100
                        cardapio.update({n_prato: [cardapio[n_prato][0],preco_item - (preco_item*percento)]})
                    elif desconto == 2:
                        novo_preco = float(input('Novo valor do prato (R$): '))
                        cardapio.update({n_prato: [cardapio[n_prato][0],novo_preco]})
                    salvar_cardapio()
                else:
                    ('Tente novamente.')
            except ValueError:
                print('Digite corretamente.')
    else:
        print('Fechando função...')
        return None

def menu_cliente(user):
    print(f'--- Menu Cliente ---')
    
    if email_senha[user]['pedido']:
        print('Você já fez um pedido:')
        for item in email_senha[user]['pedido']:
            print(f'- {item[0]} - R${item[1]:.2f}')

        if email_senha[user]['visto'] == 'Sim':
            print('Seu pedido foi visualizado.')

        escolha = input('Deseja fazer um novo pedido? (s/n): ').lower().strip()
        if escolha == 's':
            novo_pedido = pedido()  # --> Renomeado para evitar conflito com função
            if novo_pedido:
                email_senha[user]['pedido'] = novo_pedido
                salvar_dados()  # --> Uso da função para salvar
                print('Novo pedido registrado com sucesso.')
        else:
            print('Aguarde seu pedido.')
    else:
        novo_pedido = pedido()
        if novo_pedido:
            email_senha[user]['pedido'] = novo_pedido
            salvar_dados()
            print('Pedido registrado com sucesso.')

# --> Função separada para o menu de funcionários
def menu_funcionario(user):
    print(f'--- Menu Funcionários ---')
    print('''Escolha uma opção:
[0] - Sair
[1] - Gerenciar pedidos
[2] - Adicionar um item ao cardápio
[3] - Remover um item do cardápio
[4] - Alterar o preço de um produto (desconto)
''')
    
    try:
        opcao = int(input('Opção: '))
    except ValueError:
        print('Entrada inválida. Saindo...')
        return  # --> Evita crash e sai do menu

    if opcao == 1:
        gerenciar_pedidos()
    elif opcao == 2:
        adicionar_pratos()
    elif opcao == 3:
        remover_pratos()
    elif opcao == 4:
        alterar_precos()
    else:
        print('Saindo do menu de funcionário...')
```

```python
#Verifica se o cadastro foi bem sucedido
resultado = cadastro()
if resultado is None:
    print('O cadastro foi finalizado incorretamente.')
else:
    user, funcao = resultado
    print(f'Bem vindo! {user}')

    if funcao == 'Cliente':
        menu_cliente(user)  # --> Chamada da função separada para clientes
    elif funcao == 'Empregado':
        menu_funcionario(user)  # --> Chamada da função separada para funcionários
```
